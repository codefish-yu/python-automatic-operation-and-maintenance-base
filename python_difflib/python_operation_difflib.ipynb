#### 1. 比对两个差异文件

```python
import difflib

#文本１
text1 = '''this module provide class and function for compute,
           include html and text file .version7.4'''
#文本２
text2 = '''this module provide class and function for compute,
           include html and text file .version7.5'''

#使用splitlines()按照行将字符串切成列表
text1_line = text1.splitlines()
text2_line = text2.splitlines()
#创建差异比较对象
d = difflib.Differ()
diff = d.compare(text1_line,text2_line)
# print(list(diff))
print('\n'.join(list(diff)))#字符串
```

#### 2. 实战

```python
import sys
```

```python
class Filecmp:
    '''比较文件内容的类'''
    #文件内容读取方法
    def readfile(self,file_path):
        try:
            file_path = file_path
            with open(file_path,'r') as f:
                text = f.readlines()
                return text
        except:
            print('file error')
            sys.exit()
    
    #比较文件内容的方法
    def cmp(self,file1,file2):
        d = difflib.Differ()
        diff = '\n'.join(list(d.compare(file1,file2)))
        return diff

#给出需要比较的两个文件
first_filename = input('请输入您想比较的文件１：')
sec_filename = input('请输入您想比较的文件２：')
first_file_path = '/home/tarena/operation/python_difflib/{}'.format(first_filename)
sec_file_path = '/home/tarena/operation/python_difflib/{}'.format(sec_filename)

#实例化类
filecmp = Filecmp()

#读取两个文件的内容
file1 = filecmp.readfile(first_file_path)
file2 = filecmp.readfile(sec_file_path)

#比较
diff = filecmp.cmp(file1,file2)
print(diff)
```
